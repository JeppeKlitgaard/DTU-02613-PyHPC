```python
import pandas as pd
import seaborn as sns
```

```python
df1_4 = pd.read_csv("1_4_one_read.csv", delimiter="\t")
df1_5 = pd.read_csv("1_5_one_read.csv", delimiter="\t")
df1_4["NUMA"] = "Default"
df1_5["NUMA"] = "Interleave All"

df = pd.concat([df1_4, df1_5], axis=0)
df.head(3)
```

```python
speed_base = df.query("n_proc == 1")["time"].mean()
df["speedup"] = speed_base / df["time"]
```

```python
# sns.lineplot(df, x="n_proc", y="time", hue="n_run")
sns.lineplot(df, x="n_proc", y="speedup", hue="n_run", style="NUMA")

# We shold have observed Interleave All keep going up,
# something is not working.
```
